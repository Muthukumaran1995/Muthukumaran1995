### 1. Find the most expensive car company name

### 3. Find each companyâ€™s highest price car


```python
auto.groupby(['make'])['price'].max().reset_index()
```

### Sort all cars by Price column


```python
auto[['make','price']].sort_values(by=['price'], ascending=False)
```

### Group by make, fuel-type and body-style to calculate the average of city-mpg and highwaympg (miles per gallon)


```python
auto.groupby(['make','fuel_type','body_style']).agg({'city_mpg':'mean','highway_mpg':'mean'})
                         
```
